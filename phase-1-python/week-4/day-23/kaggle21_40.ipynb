```python
import numpy as np
A=np.ones(3)*1
B=np.ones(3)*2
C=np.ones(3)*3
print(f"A: {A}, B: {B}, C: {C}")
np.add(A, B,out=B)
np.divide(A,2,out=A)
np.negative(A, out=A)
np.multiply(A, B, out=A)
print(f"A: {A}, B: {B}, C: {C}")
```
