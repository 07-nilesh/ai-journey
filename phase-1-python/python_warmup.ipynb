```python
def calculate_stats_raw(numbers):
    """
    Calculates stats without using any external libraries.
    """
    if not numbers:
        return "List is empty!"

    # 1. Sort the list (needed for median)
    sorted_nums = sorted(numbers)
    n = len(numbers)

    # 2. Mean (Average)
    mean = sum(numbers) / n

    # 3. Median (Middle Value)
    if n % 2 == 0:
        # If even, average the two middle numbers
        median = (sorted_nums[n//2 - 1] + sorted_nums[n//2]) / 2
    else:
        # If odd, take the middle one
        median = sorted_nums[n//2]

    # 4. Max and Min
    maximum = max(numbers)
    minimum = min(numbers)

    return {
        "mean": mean,
        "median": median,
        "max": maximum,
        "min": minimum
    }

# Test it!
my_list = [10, 20, 30, 40, 50, 60, 70]
print(calculate_stats_raw(my_list))
          
```

```python
import numpy as np

def calculate_stats(numbers):
    """
    Takes a list of numbers and returns basic statistics.
    """
    if not numbers:
        return "List is empty!"

    stats = {
        "mean": np.mean(numbers),      # The average
        "median": np.median(numbers),  # The middle value
        "max": np.max(numbers),        # The highest value
        "min": np.min(numbers)         # The lowest value
    }
    
    return stats

# Test it out!
my_list = [10, 20, 30, 40, 50, 60, 70]
print(calculate_stats(my_list))
```

```python
import csv

def read_csv_raw(requirements):
    try:
        with open(requirements.txt, mode='r') as file:
            # Create a csv reader object
            csv_reader = csv.reader(file)
            
            # Use a loop to print only the first 5 rows
            for i, row in enumerate(csv_reader):
                if i < 5:
                    print(row)
                else:
                    break
    except FileNotFoundError:
        print("Error: Check your file path!")

# Test it with your requirements.txt (it's not a CSV, but it will print 5 lines!)
# read_csv_raw('requirements.txt')
```
