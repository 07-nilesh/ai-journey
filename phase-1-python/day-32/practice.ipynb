```python
study_hrs_list = [2.0, 4.0, 5.0]
actual_scores_list = [40.0, 70.0, 80.0]

current_weight = 12.0  # Reset weight to the exact same starting point
lr = 0.02

print("\n=== STARTING BATCH GRADIENT DESCENT ===")
for epoch in range(2):  # Run for 2 training cycles
    print(f"\n--- Epoch {epoch+1} ---")

    total_batch_gradient = 0.0  # Reset our background notepad for this epoch

    for i in range(len(study_hrs_list)):
        prediction = current_weight * study_hrs_list[i]
        loss = (prediction - actual_scores_list[i]) ** 2

        # Calculate gradient for this student
        gradient = 2 * (prediction - actual_scores_list[i]) * study_hrs_list[i]

        # ACCUMULATE: Just save the direction to our notepad. Do NOT change weight yet!
        total_batch_gradient += gradient

        print(
            f"  Student {i+1} -> Prediction: {prediction:.2f} | Loss: {loss:.2f}"
        )

    # OUTSIDE THE INNER LOOP: Calculate the average gradient of the whole batch
    average_gradient = total_batch_gradient / len(study_hrs_list)

    # SINGLE CALM UPDATE: Change the weight once per epoch using the average
    current_weight -= lr * average_gradient

    print(f"\n  [UPDATED ONCE PER EPOCH] Average Batch Gradient: {average_gradient:.2f}")
    print(f"  >>> Final Weight for Epoch {epoch+1}: {current_weight:.4f}")
```
